```python
df['quality'] = np.where(df['quality'] <= 6, 0, 1)

df['quality'].value_counts()

```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

#All all the other feature have numeric data, we can directly apply Normalization in the dataset

columns = df.columns
df[columns] = mms.fit_transform(df[columns])

df.head(10)

df.describe()

# After performing min-max scaling over the dataset,the values of all attributes squeezed between 0 and 1.
#Let us see in detail the functioning of MinMaxScaler.

# Min-Max Scaling, also known as normalization, is a technique used to transform features to a common scale.
# MinMaxScaler scales the data so that it is in the range of [0, 1]. It is useful when the distribution of the data is not Gaussian or when you want to preserve the relationships between the variables’ values.



```
